```python
import xarray as xr
files = ["RF25_ind2009_rfp25.nc", "RF25_ind2010_rfp25.nc", "RF25_ind2011_rfp25.nc"]

# Stores May–July mean
yearly_totals = []

for f in files:
    year = int(f.split('_')[1][3:])  
    ds = xr.open_dataset(f)
    rain = ds['RAINFALL']
    
    # Selects May–July months
    mjj = rain.sel(TIME=rain['TIME'].dt.month.isin([5,6,7]))
    
    # Compute mean rainfall
    P_year = float(mjj.mean(dim=['TIME','LATITUDE','LONGITUDE']).values)
    
    yearly_totals.append((year, P_year))

# Computes long-term mean across all years
P_avg = sum([v[1] for v in yearly_totals]) / len(yearly_totals)

# Computes Rainfall Index per year
print("Rainfall Index (May–July) per year:")
for year, P_year in yearly_totals:
    RI = P_year / P_avg
    print(f"{year}: {RI:.3f}")
```

```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np

# Files and years
files = ["RF25_ind2010_rfp25.nc", "RF25_ind2011_rfp25.nc", "RF25_ind2013_rfp25.nc"]
years = [2010, 2011, 2013]
months = ['June','July','August','September']

# Step 1: Compute long-term mean per month across all files (for Rainfall Index)
all_monthly_values = {6: [], 7: [], 8: [], 9: []}  # dict for June-Sept

for f in files:
    ds = xr.open_dataset(f)
    rain = ds['RAINFALL']
    for m in range(6,10):
        month_data = rain.sel(TIME=rain['TIME'].dt.month==m)
        if month_data.sizes["TIME"] > 0:
            mean_val = float(month_data.mean(dim=['TIME','LATITUDE','LONGITUDE']).values)
            all_monthly_values[m].append(mean_val)

# Long-term mean per month
long_term_mean = {m: np.mean(vals) for m, vals in all_monthly_values.items()}

# Step 2: Compute Rainfall Index per month for each year and plot
for f, year in zip(files, years):
    ds = xr.open_dataset(f)
    rain = ds['RAINFALL']
    
    # Monthly mean rainfall
    monthly_means = [float(rain.sel(TIME=rain['TIME'].dt.month==m).mean(dim=['TIME','LATITUDE','LONGITUDE']).values) for m in range(6,10)]
    
    # Rainfall Index per month
    ri = [monthly_means[i]/long_term_mean[m] for i,m in enumerate(range(6,10))]
    
    # Bar chart of Rainfall Index
    plt.figure(figsize=(6,4))
    plt.bar(months, ri, color='coral')
    plt.title(f'Rainfall Index (June–Sept) {year}')
    plt.ylabel('Rainfall Index')
    plt.show()
```

```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np

# All file groups combined
file_groups = [
    ["RF25_ind2009_rfp25.nc", "RF25_ind2010_rfp25.nc", "RF25_ind2011_rfp25.nc"],
    ["RF25_ind2010_rfp25.nc", "RF25_ind2011_rfp25.nc", "RF25_ind2012_rfp25.nc"],
    ["RF25_ind2012_rfp25.nc", "RF25_ind2013_rfp25.nc", "RF25_ind2014_rfp25.nc"]
]

all_yearly_totals = []

# Process each group
for files in file_groups:
    yearly_totals = []
    for f in files:
        year = int(f.split('_')[1][3:])
        ds = xr.open_dataset(f)
        rain = ds['RAINFALL']
        
        # Select June–September
        js = rain.sel(TIME=rain['TIME'].dt.month.isin([6,7,8,9]))
        
        # Compute mean rainfall
        P_year = float(js.mean(dim=['TIME','LATITUDE','LONGITUDE']).values)
        yearly_totals.append((year, P_year))
        all_yearly_totals.append((year, P_year))
    
    # Compute group-wise long-term mean
    P_avg = np.mean([v[1] for v in yearly_totals])
    
    # Print Rainfall Index for this group
    print("Rainfall Index (June–Sept) for group:")
    for year, P_year in yearly_totals:
        RI = P_year / P_avg
        print(f"{year}: {RI:.3f}")
    print("\n")

# --- Visualization: Bar chart of all years together ---
all_years = [v[0] for v in all_yearly_totals]
all_RI = []

# Compute overall long-term mean across all years
overall_avg = np.mean([v[1] for v in all_yearly_totals])

for year, P_year in all_yearly_totals:
    RI = P_year / overall_avg
    all_RI.append(RI)

plt.figure(figsize=(10,5))
plt.bar([str(y) for y in all_years], all_RI, color='skyblue')
plt.ylabel('Rainfall Index (June–Sept)')
plt.title('Rainfall Index per Year')
plt.show()
```
